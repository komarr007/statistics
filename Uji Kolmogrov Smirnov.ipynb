### Menghitung nilai frekuensi kumulatif dan frekuensi proporsi

```python
#sorting the data
data = data.sort_values(['Penghasilan Ayah'])
```

```python
counter = dict(collections.Counter(data['Penghasilan Ayah']))
```

```python
#get the unique data
data_baru = pd.DataFrame(counter.items(), columns=['Penghasilan','frequency'])
data = data_baru
```

```python
#membuat column frequency cumulative
data['f kumulatif'] = data['frequency'].cumsum()
```

```python
#membuat column kumulatif proporsi dengan rumus f kumulatif / n
data['kp'] = data['f kumulatif']/50
```

```python
#membuat column Z dengan rumus (Xi - mean) / sd
data['Z'] = (data['Penghasilan'] - 3119932.48)/data['Penghasilan'].std()
data['Z'] = data['Z'].round(2)
```

## Menghitung nilai Kolmogorov-smirnov table

**Jumlah data = 50, nilai alfa = 0.05**

*rumus untuk n > 35 dengan alfa = 0.05* adalah $1.36 / sqrt(n)$

```python
import math
ks_table = 1.36/math.sqrt(50)
print("Kolmogorov-smirnov table = ",ks_table)
```

H0 **ditolak** jika *nilai kolmogorov-smirnov hitung >= nilai kolmogorov-smirnov table* <br>
H0 **diterima** jika *nilai kolmogorov-smirnov hitung < nilai kolmogorov-smirnov table*

**Hipotesis :** <br><li>*H0 = Data berdistribusi normal*</li><br><li>*H1 = Data tidak berdistribusi normal*</li>

*Nilai Kolmogorov-smirnov hitung >= Nilai Kolmogorov table*<br>**0.12540 < 0.1923 <br> Maka H0 diterima**
